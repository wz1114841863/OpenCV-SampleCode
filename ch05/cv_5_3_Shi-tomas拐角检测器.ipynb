```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img25 = cv.imread('../Images/img25.jpg')
gray = cv.cvtColor(img25,cv.COLOR_BGR2GRAY)
corners = cv.goodFeaturesToTrack(gray,15,0.01,10)
corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv.circle(img25,(x,y),3,255,-1)
plt.imshow(img25),plt.show()
```

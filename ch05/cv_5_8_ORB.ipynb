```python
#ORB比SURF快得多，而SIFT和ORB描述符比SURF更好。
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img20 = cv.imread('../Images/img20.png',0)
# 初始化ORB检测器
orb = cv.ORB_create()
# 用ORB寻找关键点
kp = orb.detect(img20,None)
# 用ORB计算描述符
kp, des = orb.compute(img20, kp)
# 仅绘制关键点的位置，而不绘制大小和方向
result = cv.drawKeypoints(img20, kp, None, color=(0,255,0), flags=0)
plt.imshow(result), plt.show()
```

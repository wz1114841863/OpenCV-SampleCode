```python
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt

img12 = cv.imread("./Images/img12.jpg")
img12gray = cv.cvtColor(img12, cv.COLOR_BGR2GRAY)

ret, thresh = cv.threshold(img12gray, 10, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

plt.imshow(thresh,'gray'),plt.title("binary image")
plt.show()
print(len(contours))
print(len(hierarchy))
```

```python
#
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
#2D
img12 = cv.imread("./Images/img12.jpg")
img12 = img12[:,:,:: -1]
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img12, -1, kernel)
plt.subplot(121),plt.imshow(img12),plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging'),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#
#
img12 = cv.imread("./Images/img12.jpg")
img12 = img12[:,:,:: -1]
blur = cv.blur(img12, (7,7))
plt.subplot(121),plt.imshow(img12),plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred'),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#
#
img12 = cv.imread("./Images/img12.jpg")
img12 = img12[:,:,:: -1]
gauBlur = cv.GaussianBlur(img12, (7,7), 0)
plt.subplot(121),plt.imshow(img12),plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(gauBlur),plt.title('GaussianBlurred'),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#
#
img13 = cv.imread("./Images/img13.png")
img13 = img13[:,:,:: -1]
medBlur = cv.medianBlur(img13, 11)
plt.subplot(121),plt.imshow(img13),plt.title('Original'),plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(medBlur),plt.title('MedianBlurred'),plt.xticks([]), plt.yticks([])
plt.show()
```

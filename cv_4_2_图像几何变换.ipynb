```python
#函数：cv.getPerspectiveTransform(), cv.wrapAffine(), cv.wrapPerspectove()

import numpy as np
import cv2 as cv 
from matplotlib import pyplot as plt
#Scaling
img = cv.imread("./Images/img1.jpg")
img = img[:,:,:: -1]
res = cv.resize(img, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC)

plt.subplot(121),plt.imshow(img),plt.title("Origin Image")
plt.subplot(122),plt.imshow(res),plt.title("Expanded Image")
```

```python
#平移
rows, cols, channels = img.shape
M = np.float32([[1, 0, 200], [0, 1, 250]])
dst = cv.warpAffine(img, M, (cols, rows))
#cv.warpAffine函数的第三个参数是输出图像的大小，其形式应为(width，height)。记住width =列数，height =行数。
plt.subplot(121),plt.imshow(img),plt.title("Origin Image")
plt.subplot(122),plt.imshow(dst),plt.title("Translation Image")
```

```python
#Rotation:
# cols-1 和 rows-1 是坐标限制
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title("Origin Image")
plt.subplot(122),plt.imshow(dst),plt.title("Rotation Image")
```

```python
#仿射变换

img9 = cv.imread("./Images/img9.png")
rows, cols, channels = img9.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1, pts2)
dst = cv.warpAffine(img9, M, (cols, rows))
plt.subplot(121),plt.imshow(img9),plt.title("Origin Image")
plt.subplot(122),plt.imshow(dst),plt.title("Affine Transformation Image")
```

```python
#投射变换
img9 = cv.imread("./Images/img9.png")
rows, cols, channels = img9.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[500,0],[0,500],[500,500]])
M = cv.getPerspectiveTransform(pts1, pts2)
dst = cv.warpPerspective(img9,M,(500,500))
plt.subplot(121),plt.imshow(img9),plt.title("Origin Image")
plt.subplot(122),plt.imshow(dst),plt.title("Perspective Transformation Image")
```

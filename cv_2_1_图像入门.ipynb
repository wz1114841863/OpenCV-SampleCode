```python
import numpy as np
import cv2 as cv
import sys

img = cv.imread(cv.samples.findFile("./Images/img1.jpg"))
#cv.imread() 包含两个参数，路径和读取方式。
if img is None:
    sys.exit("Cloud not load image.")
#新建窗口，可调整大小 参数为窗口名称和标志位。
cv.namedWindow("Img", cv.WINDOW_NORMAL)
#cv.imshow() 包含两个参数，窗口标题和打开对象。
cv.imshow("Img", img)
k = cv.waitKey(0)

#cv.imwrite() 包含两个参数，保存图片名称格式和保存对象。
if k == ord("s"):
    cv.imwrite("img1_1.png", img)

cv.destroyAllWindows()
```

```python
#使用MatPlotlib
from matplotlib import pyplot as plt
#OpenCV中的图片为BGR，而matplotlib中使用RGB，因此需要颠倒顺序
img = img[:,:,::-1]
plt.imshow(img, cmap="gray", interpolation="bicubic")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#查找、绘制和分析直方图。
#直方图是针对灰度图片的
#函数：cv.calcHist(),参数：BINS, DIMS, RANGE
#cv.calcHist（images，channels，mask，histSize，ranges [，hist [，accumulate]]）
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
#计算直方图
img5 = cv.imread('../Images/img5.png',0)
hist = cv.calcHist([img5],[0],None,[256],[0,256])
"""
#使用Numpy计算
hist,nbins = np.histogram(img5, None, 256. [0,256])
"""
#绘制直方图
#使用matplotlib直接绘制
plt.subplot(1,2,1), plt.imshow(img5, 'gray'), plt.title("Origin image")
plt.subplot(1,2,2), plt.hist(img5.ravel(), 256, [0,256]),plt.title("Hist")
#plt.hist(hist)
plt.show()
```

```python
#绘制彩色图片的直方图
img19 = cv.imread("../Images/img19.jpg")
img19 = img19[:,:,:: -1]
color = ('r', 'g', 'b')
for i,col in enumerate(color):
    hist = cv.calcHist([img19], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
```

```python
# create a mask
mask = np.zeros(img19.shape[:2], np.uint8)
mask[20:200, 20:200] = 255
masked_img19 = cv.bitwise_and(img19,img19,mask = mask)
# 计算掩码区域和非掩码区域的直方图
# 检查作为掩码的第三个参数
hist_full = cv.calcHist([img19],[0],None,[256],[0,256])
hist_mask = cv.calcHist([img19],[0],mask,[256],[0,256])
plt.subplot(221), plt.imshow(img19)
plt.subplot(222), plt.imshow(mask,'gray')
plt.subplot(223), plt.imshow(masked_img19)
plt.subplot(224), plt.plot(hist_full), plt.plot(hist_mask)
plt.xlim([0,256])
plt.show()
```

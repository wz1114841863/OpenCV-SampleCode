```python
#二维直方图：两个特征，每个像素的色相和饱和度值
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
#BGR -> HSV
#channel = [0,1]，因为我们需要同时处理H和S平面。
#bins = [180,256] 对于H平面为180，对于S平面为256。
#range = [0,180,0,256] 色相值介于0和180之间，饱和度介于0和256之间。
img19 = cv.imread("../Images/img19.jpg")
hsv_img19 = cv.cvtColor(img19,cv.COLOR_BGR2HSV)   
hist = cv.calcHist([hsv_img19], [0,1], None, [180,256], [0, 180, 0, 256])
"""
cv.imshow("image", hist)
cv.waitKey(0)
cv.destroyAllWindows()
"""
img19 = img19[:,:,::-1]
plt.subplot(1,2,1),plt.imshow(img19)
plt.subplot(1,2,2),plt.imshow(hist,interpolation = 'nearest')
#X轴显示S值，Y轴显示色相
plt.show() 
```

```python
#cv.pyrUp(), cv.pyrDown()
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt

img1 = cv.imread("./Images/img1.jpg")
img1 = img1[:,:,:: -1]
imgPyr = img1.copy()

#
gaussPyr = [imgPyr]
for i in range(6):
    imgPyr = cv.pyrDown(imgPyr)
    gaussPyr.append(imgPyr)

for i in range(6):
    plt.subplot(3,2,i+1),plt.imshow(gaussPyr[i])
    
plt.show()

print(gaussPyr[4].shape)
GE = cv.pyrUp(gaussPyr[5])
print(GE.shape)
```

```python
#
lapPyr = [gaussPyr[5]]
tmp = [gaussPyr[5]]
for i in range(5, 0, -1):
    GE = cv.pyrUp(gaussPyr[i])
    tmp.append(GE)
    lap = cv.subtract(gaussPyr[i-1], GE)
    lapPyr.append(lap)

    
for i in range(6):
    plt.subplot(3,2,i+1),plt.imshow(tmp[i])

plt.show()

for i in range(6):
    plt.subplot(3,2,i+1),plt.imshow(lapPyr[i])

plt.show()
```

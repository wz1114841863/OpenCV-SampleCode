```python
#函数：cv.reode(), cv.dilate, cv.morphologyEx()
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt

img4 = cv.imread("./Images/img4.png",0)
#Erosion
kernel = np.ones((5,5), np.uint8)
erosion = cv.erode(img4, kernel, iterations = 1)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(erosion,"gray"),plt.title("Erosed image")
```

```python
#Dilation
kernel = np.ones((5,5), np.uint8)
dilation = cv.dilate(img4, kernel, iterations = 1)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(dilation,"gray"),plt.title("Dilated image")
```

```python
#Opening:erosion followed by dilation
kernel = np.ones((5,5), np.uint8)
opening = cv.morphologyEx(img4, cv.MORPH_OPEN, kernel)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(opening,"gray"),plt.title("Opening image")
```

```python
#Closing:Dilation followed by Erosion
kernel = np.ones((5,5), np.uint8)
closing = cv.morphologyEx(img4, cv.MORPH_CLOSE, kernel)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(closing,"gray"),plt.title("Closing image")
```

```python
#形态学梯度
kernel = np.ones((5,5), np.uint8)
gradient = cv.morphologyEx(img4, cv.MORPH_GRADIENT, kernel)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(gradient,"gray"),plt.title("Gradient image")
```

```python
#Top Hat
kernel = np.ones((9,9), np.uint8)
topHat = cv.morphologyEx(img4, cv.MORPH_TOPHAT, kernel)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(topHat,"gray"),plt.title("Top Hat")
```

```python
#Black Hat
kernel = np.ones((9,9), np.uint8)
blackHat = cv.morphologyEx(img4, cv.MORPH_BLACKHAT, kernel)
plt.subplot(121),plt.imshow(img4,"gray"),plt.title("Original image")
plt.subplot(122),plt.imshow(blackHat,"gray"),plt.title("Black Hat")
```

```python
#结构元素：使用cv.getStructingElement()获得所需内核
#矩形内核：
kernel1 = cv.getStructuringElement(cv.MORPH_RECT, (5,5))
print(kernel1)
#椭圆内核：
kernel2 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5,5))
print(kernel2)
#十字内核：
kernel3 = cv.getStructuringElement(cv.MORPH_CROSS, (5,5))
print(kernel3)
```

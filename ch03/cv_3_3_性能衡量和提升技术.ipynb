```python
#衡量代码的性能。cv.getTickCount，cv.getTickFrequency
#python：time模块，profile模块
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
t1 = cv.getTickCount()
#执行代码
img1 = cv.imread("../Images/img1.jpg")
img1 = img1[:,:,::-1]
plt.imshow(img1),plt.title("Img1")
plt.show()
#结束
t2 = cv.getTickCount()
time = (t2 - t1)/ cv.getTickFrequency()

print(time)
```

```python
#查看/开启优化
#函数：cv.useOptimized(),cv.setUseOptimized()
# check if optimization is enabled
cv.useOptimized()
%timeit res = cv.medianBlur(img1,49)
#disabled
cv.setUseOptimized(False)
cv.useOptimized()
%timeit res = cv.medianBlur(img1,49)
#abled
cv.setUseOptimized(True)
```

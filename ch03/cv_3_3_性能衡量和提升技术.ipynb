```python
#cv.getTickCountcv.getTickFrequency
#pythontimeprofile
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
t1 = cv.getTickCount()
#
img1 = cv.imread("../Images/img1.jpg")
img1 = img1[:,:,::-1]
plt.imshow(img1),plt.title("Img1")
plt.show()
#
t2 = cv.getTickCount()
time = (t2 - t1)/ cv.getTickFrequency()

print(time)
```
